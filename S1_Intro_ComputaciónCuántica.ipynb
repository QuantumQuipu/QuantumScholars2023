## 1. Qubit


La información clásica puede ser almacenada en un sistema físico en estados clásicos , que son estados bien determinados y finitos.

Ejemplos:

1. En un bit clásico los estados clásicos pueden ser 1 y 0
2. Un dado puede tener 6 estados por lo que  estados clásicos pueden ser 1, 2, 3,4, 5 y 6.
3. Un ventilador electrico puede tener por ejemplo 3 estados ALTO, BAJO o MEDIO

La información cuántica puede ser almacenada en 2 estados medibles (0 y 1)  y además puede encontrarse en una superposición de estos estados esto es denominado un bit cuántico es decir un qubit, esto puede ser representado usando la notación Brac-ket, con el vector de estado $ |\Psi \rangle $.

$$ | \Psi \rangle  = \alpha | 0 \rangle + \beta | 1 \rangle$$

Donde $\alpha$ y $\beta$ son números complejos denominados amplitudes.

Además los vectores $|0\rangle $ y $|1\rangle $ pueden ser representados como vectores columna

$$ | 0 \rangle = 
\begin{pmatrix}
1 \\
0 
\end{pmatrix}
$$

$$ | 1 \rangle = 
\begin{pmatrix}
0 \\
1 
\end{pmatrix}
$$

Por lo tanto el vector $ |\Psi \rangle $ toma la forma

$$ | \Psi \rangle  = \begin{pmatrix}
\alpha \\
\beta 
\end{pmatrix}$$

La norma al cuadrado de las amplitudes representan la probabilidad de medir su respectivo estado

$$Probabilidad(| 0\rangle ) = | \alpha |^{2}$$

$$Probabilidad(| 1\rangle ) = | \beta |^{2}$$

Como la probabilidad de observar alguno de los estados es segura (100%) se debe cumplir 

$$ | \alpha |^{2} +  | \beta |^{2} = 1$$

Se dice por tanto que el estado $ |\Psi \rangle $ está normalizado

## 4. Dirac Bra

Para cada vector columna existe $ |\Psi \rangle $ un vector fila $ \langle \Psi |$ el cual es el transpuesto conjugado del vector columna

$$ \langle \Psi |  = | \Psi \rangle ^{\dagger} $$ 

En la notación de Dirac estos vectores fila son denominados Bras.

Por ejemplo obtengamos el transpuesto conjugado del estado visto anteriormente $|\Phi \rangle$

$$  | \Phi \rangle = \frac{1 +2i}{3} | 0 \rangle - \frac{2}{3} | 1\rangle $$

Obtenemos

$$  \langle \Phi | = \frac{1  - 2i}{3}  \langle  0 | - \frac{2}{3} \langle  1 | $$

Estos Bras también pueden ser representados en qiskit, por ejemplo la transpuesta conjugada del vector $| 0 \rangle$ 

```python
Bra0 = [1,0]
array_to_latex(Bra0)
```

## 5. Producto Interno

Un bra en general puede representarse de la siguiente manera 

$$ \langle \Psi | = \alpha^{\ast} \langle 0 | + \beta^{\ast} \langle 1 | = ( \alpha^{\ast}, \beta^{\ast})$$

Esto nos permite definir el producto interno entre vectores de estado, el cual esta dado por:

$$\langle \Psi ||| \Psi \rangle = \langle \Psi | \Psi \rangle$$

Esto permite definir la ortogonalidad entre 2 dos vectores de estado 

$$\langle \Psi_{1} | \Psi_{2} \rangle = 0 $$

y por tanto la ortonomalidad
$$\langle \Psi_{i} | \Psi_{j} \rangle = \delta_{ij} $$

Las bases de un qubit son ortonomales, con esta condición es suficiente para definir el espacio de Hilbert de los qubits


Por tanto podemos llamar a :
$ |  0 \rangle$ y $ |  1 \rangle$  bases del qubit, ya que cumplen ambas condiciones

Y podemos verificarlo haciendo uso de numpy, con el método .dot() entre dos objetos de la clase operator en qiskit


```python
from qiskit.quantum_info.operators import Operator
```

```python
Operador_ket_0 = Operator(ket0)
Operador_ket_0
```

```python
Operador_Bra_1 = Operator(Bra1)
Operador_Bra_1
```

```python
producto_interno_10 = np.dot(Operador_Bra_1, Operador_ket_0)
array_to_latex(producto_interno_10)
```

Igualmente a la inversa podemos calcular el producto $ \langle 0 | 1 \rangle$

```python
Operador_Bra_0 = Operator(Bra0)
Operador_Bra_0
```

```python
Operador_ket_1 = Operator(ket1)
Operador_ket_1
```

```python
producto_interno_01 = np.dot(Operador_Bra_1, Operador_ket_0)
array_to_latex(producto_interno_01)
```

## 6. Producto Externo

De igual forma se puede definir otro producto, el producto externo 

$$ | \Psi \rangle \langle \phi |  = \begin{pmatrix}
\alpha  \\
\beta
\end{pmatrix} \otimes (\delta^{\ast}, \epsilon^{\ast}) = \begin{pmatrix}
\alpha\delta^{\ast}  && \alpha\epsilon^{\ast}\\
\beta\delta^{\ast} && \beta\epsilon^{\ast}
\end{pmatrix}$$

El caso especial cuando $| \Psi \rangle \langle \Psi |$ se denomina operador de proyección y permite proyectar cualquier estado en el estado $|\Psi \rangle$

Por ejemplo proyectemos el qubit general en la base $|  0 \rangle$, usando el operador de proyección $|  0 \rangle \langle 0 |$

$$|  0 \rangle \langle 0 | \Psi \rangle = |  0 \rangle \langle 0 | ( \alpha | 0 \rangle + \beta | 1 \rangle ) = \alpha |  0 \rangle \langle 0 | 0 \rangle + \beta |  0 \rangle \langle 0 | 1 \rangle = \alpha |  0 \rangle $$

por ello $\alpha = \langle 0 | \Psi \rangle$ y $\beta = \langle 1 | \Psi \rangle$

Un operador proyección importante es la indentidad

$$ |  0 \rangle \langle 0 |  + |  1 \rangle \langle 1 | = I$$

Este se obtiene de :


$$ | 0 \rangle \langle 0 |  = \begin{pmatrix}
1  \\
0
\end{pmatrix} \otimes (1, 0) = \begin{pmatrix}
1 * 1  && 1*0\\
0 * 1 &&  0*0
\end{pmatrix}= \begin{pmatrix}
1   && 0\\
0  &&  0
\end{pmatrix}$$ 

$$ | 1 \rangle \langle 1 |  = \begin{pmatrix}
0  \\
1
\end{pmatrix} \otimes (0, 1) = \begin{pmatrix}
0 * 0  && 0*1\\
1 * 0 &&  1*1
\end{pmatrix}= \begin{pmatrix}
0   && 0\\
0  &&  1
\end{pmatrix}$$ 

$$| 0 \rangle \langle 0 | + | 1 \rangle \langle 1 | = \begin{pmatrix}
1   && 0\\
0  &&  1
\end{pmatrix} = I$$   

Todo esto lo podemos hacer también en qiskit, encontremos el producto externo con la clase operator de qiskit y el método .outer() de numpy 

$$ | 0 \rangle \langle 0 | $$ 

```python
Operador_Bra_0
Operador_ket_0
Producto_externo_00 = np.outer(Operador_ket_0, Operador_ket_0)
array_to_latex(Producto_externo_00)
```

## 7. Esfera de Bloch

Podemos representar cada qubit individual (y de estado puro) como un punto en una esfera unitaria conocida como la esfera de bloch, así una representación equivalente al qubit general
        $$ | \Psi \rangle  = \alpha | 0 \rangle + \beta | 1 \rangle$$

Es :

$$ | \Psi \rangle  = \cos(\frac{\theta}{2}) | 0 \rangle + e^{i\phi} \sin(\frac{\theta}{2}) | 1 \rangle  $$ 


Esta  nueva representación hace un mapeo del bidimensional espacio de hilbert (para un sistema aislado y cerrado)definido por la base computacional y su producto interno a una esfera tridimensional

Esta fórmula puede ser derivada de 2 formas distintas, en la primera podemos considerar la forma polar de las amplitudes $\alpha$ y $\beta$

$$\alpha = r_{\alpha}e^{\phi_{\alpha} i}$$

$$\beta = r_{\beta}e^{\phi_{\beta} i}$$

Entonces el vector de estado del qubit queda

  $$ | \Psi \rangle  = r_{\alpha}e^{\phi_{\alpha} i} | 0 \rangle +  r_{\beta}e^{\phi_{\beta} i} | 1 \rangle$$

Luego podemos multiplicar el vector de estado completo por $e^{-\phi_{\alpha} i}$

 $$ e^{-\phi_{\alpha} i}| \Psi \rangle  = e^{-\phi_{\alpha} i}(r_{\alpha}e^{\phi_{\alpha} i} | 0 \rangle +  r_{\beta}e^{\phi_{\beta} i} | 1 \rangle)$$

 $$ e^{-\phi_{\alpha} i}| \Psi \rangle  = r_{\alpha} | 0 \rangle +  r_{\beta}e^{(\phi_{\beta} -\phi_{\alpha})i} | 1 \rangle$$

Este vector de estado es equivalente al primero debido a que la fase global no tiene consecuencias medibles, es así que el vector de estado original queda

 $$ | \Psi \rangle  \equiv r_{\alpha} | 0 \rangle +  r_{\beta}e^{(\phi_{\beta} -\phi_{\alpha})i} | 1 \rangle$$

como el vector de estado esta normalizado se cumple que 

$$r_{\alpha}^{2} + r_{\beta}^{2} = 1$$

por tanto ambos pueden ser parametrizados en coordenadas polares 

$$r_{\alpha} = \cos(\gamma)$$

$$r_{\beta} = \sin(\gamma)$$

y       $$\phi = \phi_{\beta} - \phi_{\alpha}$$

Por tanto

 $$ | \Psi \rangle  \equiv \cos(\gamma) | 0 \rangle +  \sin(\gamma)e^{i \phi } | 1 \rangle$$

Podemos proyectar el vector de estado en los ejes cartesianos y obtener las coordenadas cartesianas haciendo uso de las matrices de Pauli

Para el eje $x$

$$\sigma_{x}  = 
\begin{pmatrix}
0   && 1\\
1   &&  0
\end{pmatrix}
$$

Para el eje $y$

$$\sigma_{y}  = 
\begin{pmatrix}
0   && -i\\
i   &&  0
\end{pmatrix}
$$

Para el eje $z$

$$\sigma_{z}  = 
\begin{pmatrix}
1   && 0\\
0   && -1
\end{pmatrix}
$$

es así que las coordenadas cartesianas estan definidas de la siguiente forma

$$x = \langle \Psi | \sigma_{x}|\Psi \rangle$$

$$y = \langle \Psi | \sigma_{y}|\Psi \rangle$$

$$z = \langle \Psi | \sigma_{z}|\Psi \rangle$$

En el ultimo caso obtenemos

$$ z = (\cos(\gamma), \sin(\gamma)e^{i\phi}) \begin{pmatrix}
1   && 0\\
0   && -1
\end{pmatrix} \begin{pmatrix}
\cos(\gamma) \\
\sin(\gamma)e^{i\phi} 
\end{pmatrix} = \cos(\gamma)^{2} - \sin(\gamma)^{2} = \cos(2\gamma)$$

Pero en coordenadas esféricas 

$$z = \cos(\theta)$$

Por tanto, con el fin de reconciliar ambos enfoques

$$2\gamma = \theta$$

$$ \gamma = \theta /2 $$

Por tanto el ángulo polar del vector de estado es la mitad del angúlo polar correspondiente en la esfera tridimensional y obtenemos

 $$ | \Psi \rangle  \equiv \cos(\frac{\theta}{2}) | 0 \rangle +  \sin(\frac{\theta}{2})e^{i \phi } | 1 \rangle$$

En coordenadas esféricas las coordenadas $\theta$ y $\phi$ toman los siguientes valores  :
$$   0 \leq \theta \leq \pi, 0 \leq \phi \leq 2\pi $$ 

la coordenada $\theta$ determina la probabilidad de medir uno de las estados ortogonales de la base computacional es decir $| 0 \rangle$ ó $| 1 \rangle$ 

$$P(|0\rangle) = \cos(\frac{\theta}{2})^{2}$$

$$P(|1\rangle) = \sin(\frac{\theta}{2})^{2}$$

y la coordenada $\phi$  representa la fase relativa

El vector de posición del punto que representa al qubit $\vec{r}$ en coordenadas esfericas esta dado por :

$$\vec{r} = \begin{pmatrix}
\sin(\theta)\cos(\phi) \\
\sin(\theta)\sin(\phi) \\
\cos(\theta) 
\end{pmatrix}$$

y es denominado vector de Bloch

¿Cómo graficamos un vector de estado arbitrario entonces?: Debemos obtener primero su vector de Bloch

recordemos entonces un ejemplo anterior , el siguiente vector de estado

$$  | \Phi \rangle = \frac{1 +2i}{3} | 0 \rangle - \frac{2}{3} | 1\rangle $$

Escribimos el ket en qiskit

```python
ket = [[(1+2j)/3], [-2/3]]
Ket = Statevector(ket)
Ket.draw('Latex')
```

Podemos definir entonces una función que encuentre la forma polar de las 2 amplitudes y la fase relativa y así encuentre los valores de $\theta$ y $\phi$

```python
def obtener_coordenadas_esfericas(vector_de_estado):
    r_alfa = np.abs(vector_de_estado[0])
    phi_alfa = np.angle(vector_de_estado[0])
    
    r_beta = np.abs(vector_de_estado[1])
    phi_beta = np.angle(vector_de_estado[1])
    
    r = np.sqrt(r_alfa**2 + r_beta**2) # por si el vector de estado no está normalizado
    phi = phi_beta - phi_alfa
    theta = 2*np.arccos(r_alfa/r)
    return [r, theta, phi]
```

Ahora podemos usar esta función para obtener la representación del vector de estado en la esfera de Bloch

```python
plot_bloch_vector(obtener_coordenadas_esfericas(Ket), coord_type = 'spherical')
```

## 8. Estados de 2 qubits

El estado total de un sistema compuesto por 2 qubits no interactuantes puede ser representado por un único ket 

En general el estado de un sistema 2 qubits no interactuantes es una superposición de  4  estados clásicos

$$| \Psi \rangle = \alpha_{00}|00\rangle + \alpha_{01}|01\rangle + \alpha_{10}|10\rangle + \alpha_{11}|11\rangle$$

<center>
<img src="https://raw.githubusercontent.com/tonymilla/QuantumScholars2023/main/imagenes/sesion1/2qubits.png">

Como se puede ver en la imagen el estado total de un sistema compuesto por 2 qubits no interactuantes puede colapsar en cualquiera de las 4 bases con la amplitud de probabilidad correspondiente, al igual que en el caso de un solo qubit

Donde $\alpha_{00}$,$\alpha_{01}$, $\alpha_{10}$, $\alpha_{11}$ son las nuevas amplitudes de probabilidad

En una manera más formal tenemos dos sistemas independientes que se encuentran en una susperposicion de las bases $| 0 \rangle$ y $| 1 \rangle$ 

El sistema 1 esta en el estado 

$$\alpha_{1}| 0 \rangle_{1} + \beta_{1} | 1 \rangle_{1} $$

El sistema 2 está en el estado

$$\alpha_{2}| 0 \rangle_{2} + \beta_{2} | 1 \rangle_{2} $$

Entonces el estado del sistema compuesto por ambos sistemas no interactuantes esta dado en general por el producto de ambos estados

$$|\Psi\rangle = (\alpha_{1}| 0 \rangle_{1} + \beta_{1} | 1 \rangle_{1}) (\alpha_{2}| 0 \rangle_{2} + \beta_{2} | 1 \rangle_{2})$$

$$|\Psi\rangle = \alpha_{1}\alpha_{2}| 0 \rangle_{1}| 0 \rangle_{2} + \alpha_{1}\beta_{2}| 0 \rangle_{1}| 1 \rangle_{2} + \beta_{1}\alpha_{2} | 1 \rangle_{1}| 0 \rangle_{2} + \beta_{1}\beta_{2}| 1 \rangle_{1} |1\rangle_{2}$$

Tomando como convención que el primer digito binario pertenece al primer sistema podemos eliminar el subíndice y compactar la notación

$$|\Psi\rangle = \alpha_{00}| 0  0 \rangle + \alpha_{01}| 0 1 \rangle + \alpha_{10} | 10 \rangle + \alpha_{11}| 11\rangle$$

Donde $\alpha_{00}$,$\alpha_{01}$, $\alpha_{10}$, $\alpha_{11}$ cumplen con la condición de normalización:

$$|\alpha_{00}|^{2} + |\alpha_{01}|^{2} + |\alpha_{10}|^{2} + |\alpha_{11}|^{2} = 1$$

El producto que acabamos de efectuar es el producto tensorial entre 2 vectores de estado , este tambien puede realizarse con la forma columna de los kets de la siguiente forma

por ejemplo el primer termino de del estado $|00\rangle$ es el producto tensorial  $| 0 \rangle \otimes | 0 \rangle $ representa el producto del estado $| 0 \rangle$ del primer sistema y el estado $| 0 \rangle$ del segundo sistema

Para representar el estado de un sistema compuesto por 2 qubits en qiskit solo debemos ingresar en la clase Statevector una lista de números representando las amplitudes de probabilidad, esta lista es la transpuesta del vector de arriba por lo tanto se puede establecer una correspondencia entre la amplitud y su base

Para calcular las probabilidades parciales de que el qubit más a la derecha o más a la izquierda resulte ser 0 o 1, se procede de la siguiente manera 

$$|\Psi\rangle = (\frac{1}{2} + \frac{i}{2}) | 00 \rangle  + \frac{1}{2} | 01 \rangle + 0 | 10 \rangle -\frac{i}{2} | 11 \rangle $$

Entonces si queremos medir solo la probabilidad del primer sistema(el más a la derecha según la convención de IBM)  tenemos entonces:

$$|\Psi\rangle = (\frac{1}{2} + \frac{i}{2}) | 00 \rangle  + 0 | 10 \rangle + \frac{1}{2} | 01 \rangle -\frac{i}{2} | 11 \rangle $$

$$|\Psi\rangle = ((\frac{1}{2} + \frac{i}{2}) | 0 \rangle  + 0 | 1 \rangle)\otimes | 0 \rangle + (\frac{1}{2} | 0 \rangle -\frac{i}{2} | 1 \rangle) \otimes | 1 \rangle $$

Entonces la probabilidades parciales  estan dadas por:

$$P_{0} = | \frac{1}{2} + \frac{i}{2} |^{2} + |0|^{2} = \frac{1}{2}$$

$$P_{1} = | \frac{1}{2} |^{2} + |-\frac{i}{2}|^{2} = \frac{1}{2}$$

Lo mismo se puede hacer para medir el estado más a la izquierda 

$$|\Psi\rangle = (\frac{1}{2} + \frac{i}{2}) | 00 \rangle + \frac{1}{2} | 01 \rangle + 0 | 10 \rangle  -\frac{i}{2} | 11 \rangle $$

$$|\Psi\rangle = | 0\rangle \otimes ((\frac{1}{2} + \frac{i}{2}) | 0 \rangle +  \frac{1}{2} | 1 \rangle )+ | 1 \rangle \otimes (0 | 0 \rangle  -\frac{i}{2} | 1 \rangle )$$

$$P_{0} = | \frac{1}{2} + \frac{i}{2} |^{2} + |\frac{1}{2}|^{2} = \frac{3}{4}$$

$$P_{1} = | 0|^{2} + |-\frac{i}{2}|^{2} = \frac{1}{4}$$

En qiskit este se hace con el método probabilities_dict en la función se coloca el indice del estado que se quiere medir más a la derecha tiene indice 0 , más a la izquierda indice 1 según la convención de qiskit

Probabilidades parciales para el los estados más a la derecha(índice 0)

```python
probs_ket_psi = ket_psi.probabilities_dict([0])
print('probs: {}'.format(probs_ket_psi))
```

Probabilidades parciales para el los estados más a la izquierda(índice 1)

```python
probs_ket_psi = ket_psi.probabilities_dict([1])
print('probs: {}'.format(probs_ket_psi))
```

## 9. Estados de 3 qubits a N qubits

El estado de un sistema compuesto por 3 qubits no interactuantes esta dado por el siguiente ket

$$| \Psi \rangle = \alpha_{000}|000\rangle + \alpha_{001}|001\rangle + \alpha_{010}|010\rangle + \alpha_{011}|011\rangle + \alpha_{100}|100\rangle \\ +\alpha_{101}|101\rangle + \alpha_{110}|110\rangle + \alpha_{111}|111\rangle$$

Este ket es obtenido de manera analoga al caso de un sistema conformado por 2 qubits

Por tanto la representación en forma de vector columna es

$$|\Psi\rangle  = \begin{pmatrix}
\alpha_{000} \\
\alpha_{001}  \\
\alpha_{010} \\
\alpha_{011} \\
\alpha_{100} \\
\alpha_{101} \\
\alpha_{110} \\
\alpha_{111}
\end{pmatrix}$$

Donde los amplitudes de probabilidad $ \alpha_{000} $, $\alpha_{001}$, $\alpha_{010}$, $\alpha_{011}$, $\alpha_{100}$, $\alpha_{101}$, $\alpha_{110}$, $\alpha_{111}$ cumplen : 

$ |\alpha_{000}|^{2} + |\alpha_{001}|^{2} + |\alpha_{010}|^{2} + |\alpha_{011}|^{2} + |\alpha_{100}|^{2} + |\alpha_{101}|^{2} + |\alpha_{110}|^{2} + |\alpha_{111}|^{2} = 1$

En general para representar el estado de un sistema compuesto por n qubits independientes se necesita un vector columna de dimensión $2^{n}$ , es decir para representar un estado compuesto de n qubits se necesita un espacio de Hilbert de dimensión $2^{n}$

En general para un sistema con n qubits el estado del sistema esta dado por el siguiente ket:

$$| \Psi \rangle = \sum_{x \in \{0, 1\}^{n}} \alpha_{x} | x \rangle $$

Donde $\{0, 1\}^{n}$ es una cadena de n-bits 

Y la condición de normalización :

$$ \sum_{x \in \{0, 1\}^{n}} |\alpha_{x}|^{2} = 1$$

Representación de un sistema compuesto de multiples qubits usando qiskit

Análogo al caso de un sistema compuesto de 2 qubits independientes, se necesita introducir en la clase Statevector una lista conteniendo las amplitudes de probabilidad de las bases en el orden establecido

$ \alpha_{000} $, $\alpha_{001}$, $\alpha_{010}$, $\alpha_{011}$, $\alpha_{100}$, $\alpha_{101}$, $\alpha_{110}$, $\alpha_{111}$

```python
Ket_3_qubits = Statevector([0, 1, 1, 0, 1, 0, 0, 0]/np.sqrt(3))
Ket_3_qubits.draw('Latex')
```

Se puede verificar si el estado está normalizado con el método is_valid() de la clase Statevector()

```python
Ket_3_qubits.is_valid()
```

Las probabilidades pueden ser encontradas con el método .probabilities() de la clase Statevector

```python
probs_3_qubits = Ket_3_qubits.probabilities()
print('probs: {}'.format(probs_3_qubits))
```

Y las probabilidades parciales

```python
probs_Ket_3_qubits = Ket_3_qubits.probabilities_dict([0])
print('probs: {}'.format(probs_Ket_3_qubits))
```

```python
probs_Ket_3_qubits = Ket_3_qubits.probabilities_dict([2])
print('probs: {}'.format(probs_Ket_3_qubits))
```

Se pueden multiplicar bases conocidas usando el método .from_label() de la clase Statevector()

```python
ket_plus_por_0_plus = Statevector.from_label('-0+')
ket_plus_por_0_plus.draw('Latex')
```
